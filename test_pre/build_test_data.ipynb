```python
import os
import pandas as pd
import numpy as np
from collections import Counter
```

```python
def concat_file(file_list):
    file1 = '../test/' + str(file_list[0]) + '.csv'
    data = pd.read_csv(file1)
    data['label'] = file_list[0]
    for num in file_list[1:]:
        file = '../test/' + str(num) + '.csv'
        data_me = pd.read_csv(file)
        data_me['label'] = num
        data = data.append(data_me)
    # 删除无关列、全为nan的列
    data.reset_index(drop=True, inplace=True)
    data.drop(columns=['feature3_1', 'feature3_2', 'feature3_3', 
                        'feature3_4', 'feature3_5', 'feature3_6', 
                        'feature3_7', 'feature3_8', 'feature14', 'Date & Time'], 
                        inplace=True)
    data.dropna(axis=1, how='all', inplace=True)

    return data
```

```python
files = list(range(150))
data = concat_file(files)
data.drop(columns='feature0')
```
