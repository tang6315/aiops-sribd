```python
import numpy as np
import networkx as nx
```

```python
nodes = ['feature0', 'feature1', 'feature2', 'feature11', 'feature12',
        'feature13', 'feature15', 'feature16', 'feature17', 'feature18',
        'feature19', 'feature20_0', 'feature20_1', 'feature20_2', 'feature20_3',
        'feature20_4', 'feature20_5', 'feature20_6', 'feature20_7',
        'feature28_0', 'feature28_1', 'feature28_2', 'feature28_3',
        'feature28_4', 'feature28_5', 'feature28_6', 'feature28_7',
        'feature36_0', 'feature36_1', 'feature36_2', 'feature36_3',
        'feature36_4', 'feature36_5', 'feature36_6', 'feature36_7', 'feature60',
        'feature61_0', 'feature61_1', 'feature61_2', 'feature61_3',
        'feature61_4', 'feature61_5', 'feature61_6', 'feature61_7',
        'feature69_0', 'feature69_1', 'feature69_2', 'feature69_3',
        'feature69_4', 'feature69_5', 'feature69_6', 'feature69_7']
```

```python
DG = nx.DiGraph()
DG.add_nodes_from(nodes)
```

```python
edges_part = [(15, 0), (13, 0), (2, 0), (1, 0), 
             (12, 0), (18, 2), (18, 1), (18, 16), 
             (16, 2), (16, 1), (17, 2), (11, 1), (11, 12), 
             (19, 18), (60, 18)]
```

```python
x_n = []
for x in ['61_', '69_']:
    for num in range(8):
        for n in [1, 2, 17]:
            x_n.append((x + str(num), n))
edges_part.extend(x_n)

_20_n_17 = []
for n in range(8):
    _20_n_17.append(('20_' + str(n), 17))
edges_part.extend(_20_n_17)

y_n = []
for y in ['28_', '36_']:
    for num in range(8):
        y_n.append((y + str(num), 17))
edges_part.extend(y_n)
        
y_x = []
y_list = []
x_list = []
for y in ['28_', '36_']:
    for n in range(8):
        y_list.append(y + str(n))
for x in ['61_', '69_']:
    for n in range(8):
        x_list.append(x + str(n))
for y in y_list: 
    for x in x_list:
        if x[-1] == y[-1]:
            y_x.append((y, x))
edges_part.extend(y_x)
```

```python
DG.add_edges_from([('feature' + str(edge[0]), 'feature' + str(edge[1])) for edge in edges_part])
```

```python
nx.draw(DG, pos = nx.random_layout(DG), node_color='b', edge_color='r', with_labels=
        True, font_size=18, node_size=20)
```

```python
A = np.array(nx.adjacency_matrix(DG).todense())
```
